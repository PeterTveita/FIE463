# Print information about the environment

- This notebook prints information about your computer, the Python version, and the installed packages.
- Run this notebook to test whether the central packages required in this course are installed in your Anaconda environment.
- For each package, the code either prints the installed version, or an error message if the package was not found.

```python
from IPython.display import Markdown, display
```

***
## Information about this computer

```python
# Print some information about this computer
import platform

display(Markdown(f"<span style='font-size: 18px;'><b>Operating System:</b> {platform.system()} {platform.release()} ({platform.version()})</span>"))
display(Markdown(f"<span style='font-size: 18px;'><b>Computer name:</b> {platform.node()}</span>"))
```

***
## Python version

```python
display(Markdown(f"<span style='font-size: 18px;'><b>Python version:</b> {platform.python_version()}</span>"))
```

***
## Installed packages

```python

# Template strings for package availability messages
success_message = "<span style='font-size: 18px; color: green;'>✓ {package} {version} is installed</span>"
failure_message = "<span style='font-size: 18px; color: red;'>✗ {package} is not installed</span>"
```

```python
try:
    import numpy
    display(Markdown(success_message.format(package="NumPy", version=numpy.__version__)))
except ImportError:
    display(Markdown(failure_message.format(package="NumPy")))
```

```python
try:
    import pandas
    display(Markdown(success_message.format(package="pandas", version=pandas.__version__)))
except ImportError:
    display(Markdown(failure_message.format(package="pandas")))
```

```python
try:
    import matplotlib
    display(Markdown(success_message.format(package="matplotlib", version=matplotlib.__version__)))
except ImportError:
    display(Markdown(failure_message.format(package="matplotlib")))
```
